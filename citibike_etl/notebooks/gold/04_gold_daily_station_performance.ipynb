```python
import pyspark.sql.functions as F
```

```python
catalog = dbutils.widgets.get("catalog")
```

```python
df = spark.table('citibikes_dev.02_silver.jc_citibikes')
```

```python
df = df.groupBy('trip_date', 'start_station_name').agg(
    F.round(F.count('rider_d'),2).alias('total_rides'),
    F.round(F.avg('trip_duration_mins'),2).alias('avg_trip_duration_mins')
)
```

```python
df.write.mode('overwrite').option('overwriteSchems', 'true').saveAsTable(f'{catalog}.03_gold.daily_station_performance')
```
