```python
import pyspark.sql.functions as F
import dlt
```

```python
catalog = spark.conf.get('catalog')
```

```python
@dlt.table(
    name = "daily_summary_ride",
    comment =" this table aggregates the daily rides"
)
def transform_summary():
    df = spark.table('silver_dlt_citibikes')
    df = (df.groupBy('trip_start_date').agg(
        F.round(F.max('trip_duration_mins'),2).alias('max_trip_duration_mins'),
        F.round(F.min('trip_duration_mins'),2).alias('min_trip_duration_mins'),
        F.round(F.avg('trip_duration_mins'),2).alias('Avg_trip_duration_mins'),
        F.count('rider_d').alias('total_trips')
    ))
    return df 
```
