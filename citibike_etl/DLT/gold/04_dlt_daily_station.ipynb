```python
import pyspark.sql.functions as F
import dlt
```

```python
catalog = spark.conf.get('catalog')
```

```python
@dlt.table(
    name = 'daily_station_ride',
    comment = 'this is gold layer to calculate avg trip duration based on day and station'
)
def transform_station():
    df = spark.table('silver_dlt_citibikes')
    df = (df.groupby("trip_start_date", "start_station_name").agg(
        F.count('rider_d').alias('total_trips'),
        F.round(F.avg('trip_duration_mins'),2).alias('avg_duration')
    ))
    return df
```
