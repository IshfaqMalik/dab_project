```python
import dlt
import pyspark.sql.functions as F
```

```python
catalog = spark.conf.get("catalog")
```

```python
@dlt.table(
    name= "silver_dlt_citibikes",
    comment = 'This is cleaned data of citibikes'
)
def transform_silver():
    df = spark.table('bronze_dlt_citibikes')
    df = (df.withColumn('trip_duration_mins', (F.to_unix_timestamp(F.col('ended_at')) - F.to_unix_timestamp(F.col('started_at')))/60)
          .withColumn('trip_start_date', F.to_date(F.col('started_at')) ))
    
    return df 
```
